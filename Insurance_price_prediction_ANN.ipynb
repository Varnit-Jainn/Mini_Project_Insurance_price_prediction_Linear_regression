```python
X1=df.iloc[:,0:-1]
Y1=df.iloc[:,-1]
```

```python
X1
```

```python
Y1
```

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,Y1_train,Y1_test=train_test_split(X1,Y1,test_size=0.2,random_state=1)
```

```python
X1_train
```

```python
X1_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X1_train_scaled=scaler.fit_transform(X1_train)
X1_test_scaled=scaler.transform(X1_test)
```

```python
X1_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
X1
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
model = keras.Sequential()
model.add(Dense(8,input_dim=6,activation="relu"))
model.add(Dense(16,activation="relu"))
model.add(Dense(32,activation="relu"))
keras.layers.Dropout(0.3),
model.add(Dense(1,activation="linear"))
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer="adam")
```

```python
X1_train_scaled.shape
```

```python
Y1_train.shape
```

```python
X1_test_scaled.shape
```

```python
Y_test.shape
```

```python
history=model.fit(X1_train_scaled, Y1_train, validation_data=(X1_test_scaled, Y1_test), epochs=1000, callbacks=callback)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(X1_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(Y1_test,y_pred)
```
